

## Observations

1. The trend component indicates the long-term direction of sales over the four-year period.

3. The residual component captures random variations that cannot be explained by trend or seasonality.

4. Large residual spikes may correspond to promotional events, festivals, stock shortages, or unexpected market conditions.

Checking for stationarity


```python
# AUGUMENTED DICKEY-FULLER TEST
def adf_test(series):

    result = adfuller(series.dropna())

    print("ADF Statistic :", result[0])
    print("p-value :", result[1])
    print()

    print("Critical Values")

    for key, value in result[4].items():
        print(f"{key} : {value}")

    print()

    if result[1] <= 0.05:
        print("Result: The series is Stationary.")
    else:
        print("Result: The series is Non-Stationary.")
adf_test(monthly_sales)
```

A stationary time series has a constant mean and variance over time.
Many forecasting models, especially SARIMA, require stationary data for accurate predictions.

If the p-value is less than or equal to 0.05, the series is considered stationary.

If the p-value is greater than 0.05, the series is considered non-stationary and differencing is required.

Apply differencing

```python
monthly_sales_diff = monthly_sales.diff().dropna()
```

Re-run ADF test

```python
adf_test(monthly_sales_diff)
```

The Augmented Dickey-Fuller (ADF) Test was used to evaluate stationarity.

If the original series was found to be non-stationary, first-order differencing was applied to stabilize the data.

The processed time series is now ready for forecasting using SARIMA, Prophet, and XGBoost models.

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    mean_absolute_percentage_error
)

import numpy as np
```

# Model 1 – SARIMA (Statistical Forecasting)

## Objective

SARIMA (Seasonal AutoRegressive Integrated Moving Average) is a statistical forecasting model designed for time-series data with seasonality.

In this section, we will:

- Split the monthly sales data into training and testing sets.
- Train a SARIMA model.
- Forecast sales for the testing period.
- Evaluate the model using MAE, RMSE, and MAPE.
- Forecast the next 3 months.
- Visualize predictions with confidence intervals.

```python
sarima_model = SARIMAX(
    train,
    order=(1,1,1),
    seasonal_order=(1,1,1,12),
    enforce_stationarity=False,
    enforce_invertibility=False
)
```

The ADF Test indicated that first-order differencing was required.

Therefore:

(p,d,q) = (1,1,1)

Seasonal (P,D,Q,m)

(1,1,1,12)

were selected because:

• d=1 removes the trend

• m=12 captures yearly seasonality

• AR and MA orders are kept simple due to limited observations.

Train model

```python
sarima_fit = sarima_model.fit()

print(sarima_fit.summary())
```

predict test data

```python
sarima_pred = sarima_fit.forecast(
    steps=len(test)
)

sarima_pred
```

Metrics

```python
sarima_mae = mean_absolute_error(
    test,
    sarima_pred
)

sarima_rmse = np.sqrt(
    mean_squared_error(
        test,
        sarima_pred
    )
)

sarima_mape = mean_absolute_percentage_error(
    test,
    sarima_pred
)

print(f"MAE  : {sarima_mae:.2f}")
print(f"RMSE : {sarima_rmse:.2f}")
print(f"MAPE : {sarima_mape:.4f}")
```

3-month future forecast

```python
sarima_future = sarima_fit.get_forecast(
    steps=3
)

sarima_forecast = sarima_future.predicted_mean

confidence = sarima_future.conf_int()
```

Plot Actual vs predicted

```python
plt.figure(figsize=(14,6))

plt.plot(
    train.index,
    train.values,
    label="Training Data",
    linewidth=2
)

plt.plot(
    test.index,
    test.values,
    label="Actual Sales",
    linewidth=2
)

plt.plot(
    test.index,
    sarima_pred,
    marker="o",
    linewidth=2,
    label="SARIMA Prediction"
)

plt.title("SARIMA - Actual vs Predicted Sales")

plt.xlabel("Date")

plt.ylabel("Sales")

plt.legend()

plt.grid(True)

plt.show()
```

```python
#Display forecast
forecast_df = pd.DataFrame({

    "Forecast": sarima_forecast,

    "Lower CI": confidence.iloc[:,0],

    "Upper CI": confidence.iloc[:,1]

})

forecast_df
```

```python
#Plot Future FOrecast
plt.figure(figsize=(14,6))

plt.plot(
    monthly_sales,
    label="Historical Sales",
    linewidth=2
)

plt.plot(
    sarima_forecast.index,
    sarima_forecast.values,
    marker="o",
    linewidth=2,
    label="Future Forecast"
)

plt.fill_between(
    confidence.index,
    confidence.iloc[:,0],
    confidence.iloc[:,1],
    alpha=0.25,
    label="95% Confidence Interval"
)

plt.title("SARIMA Future Sales Forecast")

plt.xlabel("Date")

plt.ylabel("Sales")

plt.legend()

plt.grid(True)

plt.show()
```

# Model 3 – XGBoost Forecasting

## Objective

XGBoost is a machine learning regression algorithm that can also be used for time-series forecasting after converting the time series into a supervised learning dataset.

In this section we will:

- Create lag features
- Create rolling mean feature
- Extract calendar features
- Train an XGBoost Regressor
- Evaluate the model
- Forecast the next three months recursively

```python
!pip install xgboost -q
```

```python
from xgboost import XGBRegressor

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    mean_absolute_percentage_error
)

import numpy as np
```

Create time series dat

```python
xgb_df = monthly_sales.reset_index()

xgb_df.columns = ["Date", "Sales"]

xgb_df.head()
```

Create lag features

```python
# Previous month sales
xgb_df["Lag_1"] = xgb_df["Sales"].shift(1)

# Two months ago
xgb_df["Lag_2"] = xgb_df["Sales"].shift(2)

# Three months ago
xgb_df["Lag_3"] = xgb_df["Sales"].shift(3)
```

Rolling mean

```python
xgb_df["Rolling_Mean_3"] = (
    xgb_df["Sales"]
    .rolling(3)
    .mean()
)
```

Calendar features

```python
xgb_df["Month"] = xgb_df["Date"].dt.month

xgb_df["Quarter"] = xgb_df["Date"].dt.quarter

xgb_df["Year"] = xgb_df["Date"].dt.year
```

Season Feature

```python
def get_season(month):

    if month in [12,1,2]:
        return 1      # Winter

    elif month in [3,4,5]:
        return 2      # Summer

    elif month in [6,7,8]:
        return 3      # Monsoon

    else:
        return 4      # Autumn


xgb_df["Season"] = xgb_df["Month"].apply(get_season)
```

Remove Missing Values

```python
xgb_df.dropna(inplace=True)

xgb_df.reset_index(drop=True, inplace=True)

xgb_df.head()
```

Feature Matrix

```python
features = [

    "Lag_1",

    "Lag_2",

    "Lag_3",

    "Rolling_Mean_3",

    "Month",

    "Quarter",

    "Year",

    "Season"

]

X = xgb_df[features]

y = xgb_df["Sales"]
```

Train-Test Split

```python
forecast_horizon = 3

X_train = X.iloc[:-forecast_horizon]

X_test = X.iloc[-forecast_horizon:]

y_train = y.iloc[:-forecast_horizon]

y_test = y.iloc[-forecast_horizon:]
```

Train Model

```python
xgb_model = XGBRegressor(

    n_estimators=200,

    learning_rate=0.05,

    max_depth=3,

    random_state=42

)

xgb_model.fit(

    X_train,

    y_train

)
```

Predict Test Data

```python
xgb_pred = xgb_model.predict(X_test)
```

Metrics

```python
xgb_mae = mean_absolute_error(

    y_test,

    xgb_pred

)

xgb_rmse = np.sqrt(

    mean_squared_error(

        y_test,

        xgb_pred

    )

)

xgb_mape = mean_absolute_percentage_error(

    y_test,

    xgb_pred

)

print(f"MAE  : {xgb_mae:.2f}")

print(f"RMSE : {xgb_rmse:.2f}")

print(f"MAPE : {xgb_mape:.4f}")
```

Actual vs Predicted

```python
plt.figure(figsize=(14,6))

plt.plot(

    xgb_df["Date"].iloc[-forecast_horizon:],

    y_test,

    marker="o",

    linewidth=2,

    label="Actual"

)

plt.plot(

    xgb_df["Date"].iloc[-forecast_horizon:],

    xgb_pred,

    marker="o",

    linewidth=2,

    label="Predicted"

)

plt.title("XGBoost - Actual vs Predicted")

plt.xlabel("Date")

plt.ylabel("Sales")

plt.legend()

plt.grid(True)

plt.show()
```

Retrain on Full Dataset

```python
X_full = xgb_df[features]

y_full = xgb_df["Sales"]

xgb_model.fit(X_full, y_full)
```

Recursive Forecast (Next 3 Months)

```python
future_predictions = []

temp_df = xgb_df.copy()

for _ in range(3):

    last = temp_df.iloc[-1]

    next_date = last["Date"] + pd.offsets.MonthEnd(1)

    next_month = next_date.month
    next_quarter = next_date.quarter
    next_year = next_date.year
    next_season = get_season(next_month)

    row = pd.DataFrame({
        "Lag_1": [last["Sales"]],
        "Lag_2": [last["Lag_1"]],
        "Lag_3": [last["Lag_2"]],
        "Rolling_Mean_3": [[last["Sales"], last["Lag_1"], last["Lag_2"]].__iter__() and np.mean([last["Sales"], last["Lag_1"], last["Lag_2"]])],
        "Month": [next_month],
        "Quarter": [next_quarter],
        "Year": [next_year],
        "Season": [next_season]
    })

    pred = xgb_model.predict(row)[0]

    future_predictions.append(pred)

    new_row = {
        "Date": next_date,
        "Sales": pred,
        "Lag_1": last["Sales"],
        "Lag_2": last["Lag_1"],
        "Lag_3": last["Lag_2"],
        "Rolling_Mean_3": np.mean([pred, last["Sales"], last["Lag_1"]]),
        "Month": next_month,
        "Quarter": next_quarter,
        "Year": next_year,
        "Season": next_season
    }

    temp_df = pd.concat([temp_df, pd.DataFrame([new_row])], ignore_index=True)
```

Display Forecast

```python
future_dates = pd.date_range(

    monthly_sales.index[-1] + pd.offsets.MonthEnd(1),

    periods=3,

    freq="ME"

)

xgb_forecast = pd.DataFrame({

    "Date": future_dates,

    "Forecast": future_predictions

})

xgb_forecast
```

Plot future forecast

```python
plt.figure(figsize=(14,6))

plt.plot(

    monthly_sales,

    label="Historical Sales",

    linewidth=2

)

plt.plot(

    future_dates,

    future_predictions,

    marker="o",

    linewidth=2,

    label="Future Forecast"

)

plt.title("XGBoost Future Forecast")

plt.xlabel("Date")

plt.ylabel("Sales")

plt.legend()

plt.grid(True)

plt.show()
```

```python
xgb_forecast.to_csv(

    "xgboost_forecast.csv",

    index=False

)

print("Forecast saved successfully.")
```

```python
from prophet import Prophet

def prophet_forecast_segment(data, date_col, sales_col, periods=3):
    """
    Forecast future sales for a given segment using Prophet.
    """

    monthly = (
        data.groupby(pd.Grouper(key=date_col, freq="ME"))[sales_col]
        .sum()
        .reset_index()
    )

    monthly.columns = ["ds", "y"]

    model = Prophet(
        yearly_seasonality=True,
        weekly_seasonality=False,
        daily_seasonality=False
    )

    model.fit(monthly)

    future = model.make_future_dataframe(
        periods=periods,
        freq="ME"
    )

    forecast = model.predict(future)

    return forecast[["ds", "yhat"]].tail(periods)
```

Forecast Each Product Category

```python
furniture = df[df["Category"] == "Furniture"]

technology = df[df["Category"] == "Technology"]

office = df[df["Category"] == "Office Supplies"]

furniture_forecast = prophet_forecast_segment(
    furniture,
    "Order Date",
    "Sales"
)

technology_forecast = prophet_forecast_segment(
    technology,
    "Order Date",
    "Sales"
)

office_forecast = prophet_forecast_segment(
    office,
    "Order Date",
    "Sales"
)
```

Forecast Regions

```python
west = df[df["Region"] == "West"]

east = df[df["Region"] == "East"]

west_forecast = prophet_forecast_segment(
    west,
    "Order Date",
    "Sales"
)

east_forecast = prophet_forecast_segment(
    east,
    "Order Date",
    "Sales"
)
```

**TASK 5 ANOMALY DETECTION IN SALES DATA**


## Objective

The objective of this task is to identify unusual sales patterns that significantly differ from normal business behavior.

Two anomaly detection techniques are used:

1. Isolation Forest (Machine Learning)
2. Z-Score based Detection (Statistical Method)

The detected anomalies are visualized and compared to understand differences between the approaches.

```python
from sklearn.ensemble import IsolationForest

from scipy.stats import zscore
```

Weekly Sales Data

```python
weekly_sales = (
    df.groupby(
        pd.Grouper(key="Order Date", freq="W")
    )["Sales"]
    .sum()
    .reset_index()
)

weekly_sales.rename(
    columns={"Order Date":"Week"},
    inplace=True
)

weekly_sales.head()
```

```python
plt.figure(figsize=(14,5))

plt.plot(
    weekly_sales["Week"],
    weekly_sales["Sales"]
)

plt.title("Weekly Sales")

plt.xlabel("Week")

plt.ylabel("Sales")

plt.grid(True)

plt.show()
```

## Isolation Forest

Isolation Forest detects anomalies by randomly partitioning observations.

Anomalies are easier to isolate because they differ significantly from the majority of the data.

Train

```python
iso_model = IsolationForest(

    contamination=0.05,

    random_state=42

)

weekly_sales["Isolation"] = iso_model.fit_predict(

    weekly_sales[["Sales"]]

)
```

Convert Labels

```python
weekly_sales["Isolation"] = weekly_sales["Isolation"].map({

    1:"Normal",

    -1:"Anomaly"

})
```

Display Anomalies

```python
isolation_anomalies = weekly_sales[

    weekly_sales["Isolation"]=="Anomaly"

]

isolation_anomalies
```

```python
plt.figure(figsize=(15,6))

plt.plot(

    weekly_sales["Week"],

    weekly_sales["Sales"],

    label="Weekly Sales"

)

plt.scatter(

    isolation_anomalies["Week"],

    isolation_anomalies["Sales"],

    color="red",

    s=100,

    label="Anomaly"

)

plt.title("Isolation Forest - Weekly Sales Anomalies")

plt.xlabel("Week")

plt.ylabel("Sales")

plt.legend()

plt.grid(True)

plt.show()
```

```python
print("Possible explanations for detected anomalies:")

for _, row in isolation_anomalies.iterrows():

    month = row["Week"].month

    if month in [11,12]:

        reason = "Possible festive or holiday season sales."

    elif month == 1:

        reason = "Possible post-holiday demand changes."

    elif month in [6,7]:

        reason = "Possible seasonal promotion or mid-year sale."

    else:

        reason = "May be caused by promotional campaigns, inventory shortages, or unusual customer demand."

    print(

        row["Week"].date(),

        "-",

        reason

    )
```

**Z-Score Detection**
* A rolling mean is calculated to estimate expected sales.

* A week is flagged as anomalous if its sales differ by more than **2 standard deviations** from the rolling average.

Rolling mean

```python
weekly_sales["RollingMean"] = (

    weekly_sales["Sales"]

    .rolling(window=4)

    .mean()

)

weekly_sales["RollingStd"] = (

    weekly_sales["Sales"]

    .rolling(window=4)

    .std()

)
```

Calculate Z-Score

```python
weekly_sales["ZScore"] = (

    weekly_sales["Sales"]

    - weekly_sales["RollingMean"]

) / weekly_sales["RollingStd"]
```

Flag Anomalies

```python
weekly_sales["Z_Anomaly"] = (

    weekly_sales["ZScore"].abs() > 2

)
```

```python
zscore_anomalies = weekly_sales[

    weekly_sales["Z_Anomaly"]

]

zscore_anomalies
```

```python
plt.figure(figsize=(15,6))

plt.plot(

    weekly_sales["Week"],

    weekly_sales["Sales"],

    label="Weekly Sales"

)

plt.scatter(

    zscore_anomalies["Week"],

    zscore_anomalies["Sales"],

    color="green",

    s=100,

    label="Z-Score Anomaly"

)

plt.title("Z-Score Based Weekly Sales Anomalies")

plt.xlabel("Week")

plt.ylabel("Sales")

plt.legend()

plt.grid(True)

plt.show()
```

Compare Both Methods


```python
print("Isolation Forest Anomalies")

print(len(isolation_anomalies))

print()

print("Z-Score Anomalies")

print(len(zscore_anomalies))
```

Common Anomalies

```python
common = pd.merge(

    isolation_anomalies,

    zscore_anomalies,

    on="Week"

)

common[["Week"]]
```

# Observations

### Isolation Forest

- Detects anomalies using a machine learning approach.
- Can identify both global and subtle local anomalies.
- Does not assume any statistical distribution.

### Z-Score Method

- Uses the rolling mean and rolling standard deviation.
- Flags observations that deviate more than two standard deviations from the expected pattern.

**TASK 6 PRODUCT DEMAND SEGMENTATION USING CLUSTERING **



## Objective

The objective of this task is to group product sub-categories into demand segments based on their sales characteristics.

Features used:

- Total Sales Volume
- Sales Growth Rate (Year-over-Year)
- Sales Volatility
- Average Order Value

K-Means Clustering is applied to identify products with similar demand patterns.

The Elbow Method is used to determine the optimal number of clusters.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

import warnings
warnings.filterwarnings("ignore")
```

Monthly Sales Dataset

```python
monthly_product_sales = (

    df.groupby(

        [

            "Sub-Category",

            pd.Grouper(

                key="Order Date",

                freq="ME"

            )

        ]

    )["Sales"]

    .sum()

    .reset_index()

)

monthly_product_sales.head()
```

Total Sales Volume

```python
total_sales = (

    monthly_product_sales

    .groupby("Sub-Category")["Sales"]

    .sum()

    .rename("TotalSales")

)
```

Sales Growth Rate (Year-over-Year)

```python
yearly_sales = (

    df.groupby(

        [

            "Sub-Category",

            df["Order Date"].dt.year

        ]

    )["Sales"]

    .sum()

    .reset_index()

)

yearly_sales.columns = [

    "Sub-Category",

    "Year",

    "Sales"

]

yearly_sales.head()
```

Growth

```python
growth = (

    yearly_sales

    .groupby("Sub-Category")["Sales"]

    .pct_change()

)

yearly_sales["GrowthRate"] = growth

# AVERAGE GROWTH
growth_rate = (

    yearly_sales

    .groupby("Sub-Category")["GrowthRate"]

    .mean()

    .fillna(0)

)
```

Sales Volatility

```python
volatility = (

    monthly_product_sales

    .groupby("Sub-Category")["Sales"]

    .std()

    .rename("Volatility")

)
```

Average Order Value

```python
avg_order = (

    df.groupby("Sub-Category")["Sales"]

    .mean()

    .rename("AverageOrderValue")

)
```

Combine All Features

```python
cluster_df = pd.concat(

    [

        total_sales,

        growth_rate,

        volatility,

        avg_order

    ],

    axis=1

)

cluster_df.fillna(0, inplace=True)

cluster_df
```

Feature Scaling

```python
scaler = StandardScaler()

scaled_features = scaler.fit_transform(cluster_df)
```

Why scaling?

K-Means uses Euclidean distance.

Without scaling, Total Sales would dominate the clustering because it has much larger values than Growth Rate or Volatility.

**Elbow Method**

```python
inertia = []

K = range(1,11)

for k in K:

    model = KMeans(

        n_clusters=k,

        random_state=42,

        n_init=10

    )

    model.fit(scaled_features)

    inertia.append(model.inertia_)
```

Plot

```python
plt.figure(figsize=(8,5))

plt.plot(

    K,

    inertia,

    marker="o"

)

plt.xlabel("Number of Clusters")

plt.ylabel("Inertia")

plt.title("Elbow Method")

plt.grid(True)

plt.show()
```

Train K-Means

```python
kmeans = KMeans(

    n_clusters=4,

    random_state=42,

    n_init=10

)

cluster_df["Cluster"] = kmeans.fit_predict(

    scaled_features

)
```

View Cluster

```python
cluster_df
```

Cluster Summary

```python
cluster_summary = (

    cluster_df

    .groupby("Cluster")

    .mean()

)

cluster_summary
```

Cluster Labels

```python
cluster_names = {

    0:"High Volume, Stable Demand",

    1:"Growing Demand",

    2:"Low Volume, High Volatility",

    3:"Declining Demand"

}

cluster_df["Segment"] = (

    cluster_df["Cluster"]

    .map(cluster_names)

)
```

PCA for Visualization

```python
pca = PCA(

    n_components=2,

    random_state=42

)

principal_components = pca.fit_transform(

    scaled_features

)

cluster_df["PC1"] = principal_components[:,0]

cluster_df["PC2"] = principal_components[:,1]
```

Scatter Plot

```python
plt.figure(figsize=(10,7))

for segment in cluster_df["Segment"].unique():

    temp = cluster_df[

        cluster_df["Segment"]==segment

    ]

    plt.scatter(

        temp["PC1"],

        temp["PC2"],

        s=120,

        label=segment

    )

for subcat in cluster_df.index:

    plt.text(

        cluster_df.loc[subcat,"PC1"],

        cluster_df.loc[subcat,"PC2"],

        subcat,

        fontsize=8

    )

plt.title("Product Demand Segmentation")

plt.xlabel("Principal Component 1")

plt.ylabel("Principal Component 2")

plt.legend()

plt.grid(True)

plt.show()
```
